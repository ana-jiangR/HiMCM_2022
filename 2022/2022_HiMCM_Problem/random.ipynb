```python
# Modifying the Standard Deviation of a Normal Distribution
from numpy.core.defchararray import numpy
from numpy.random import normal
import matplotlib.pyplot as plt
import seaborn as sns

norm = normal(loc=10, scale=2, size=10000)
# norm = numpy.random.uniform(low=-2, high=2, size=10000)
# print(type(norm))

sns.histplot(norm, kde=True)

plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go


sample = [2,-1,2.4,0,4,100]

avg = np.average(sample)
std = np.std(sample, ddof=1)

print(avg)
print(std)

x = np.random.normal(avg, std, 1000)
fig = go.Figure(data=[go.Histogram(x=x)])
fig.show()
```
