```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go


x = np.arange(1, 10, 0.1)


yn = x
ysqrt = np.sqrt(x)
ylogn = np.log2(x)
ynlogn = x * np.log2(x)
yn2    = x * x
yploy2 = 2 * x * x - 2 * x + 1
yexp = np.exp(x)

fig = go.Figure()
fig.add_traces(go.Scatter(x=x, y=ysqrt, mode = "lines", name = "sqrt(n)"))
fig.add_traces(go.Scatter(x=x, y=ylogn, mode = "lines", name = "log2(n)"))
fig.add_traces(go.Scatter(x=x, y=yn, mode = "lines", name = "n"))
fig.add_traces(go.Scatter(x=x, y=ynlogn, mode = "lines", name = "nlog2(n)"))
fig.add_traces(go.Scatter(x=x, y=yn2, mode = "lines", name = "n^2"))
fig.add_traces(go.Scatter(x=x, y=yploy2, mode = "lines", name = "ploy2"))
fig.add_traces(go.Scatter(x=x, y=yexp, mode = "lines", name = "exp"))



fig.update_layout( 
        title='OOOO',
        title_x=0.5,
        # autosize=False,
        width=2500,
        height=1200,
    )

fig.show()
```
