```python
from cmath import nan
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_excel("LakeMead_MonthlyElevationData.xlsx", index_col=0)

MonthList = ['JAN', 'FEB',	'MAR',	'APR',	'MAY',
             'JUN',	'JUL',	'AUG',	'SEP',	'OCT',	'NOV',	'DEC']
YearStart = 1935
YearEnd = 2021

Maxi = 0
Mini = 20000
newdf = pd.DataFrame(columns=['Month', 'Year', 'Elevation'])
for year in range(YearStart, YearEnd+1):
    for mth in MonthList:
        Ele = df.at[year, mth]
        Maxi = max(Maxi, Ele)
        Mini = min(Mini, Ele)
        new_row = pd.Series({'Month': mth, 'Year': year, 'Elevation': Ele})
        newdf = pd.concat([newdf, new_row.to_frame().T], ignore_index=True)

# print(newdf)


# df = px.data.gapminder()
# df = pd.read_csv("gapminderDataFiveYear.csv")
# df = pd.read_csv("animationLakeMead.csv")

# fig = px.bar(df, x="continent", y="pop", color="continent",
#   animation_frame="year", animation_group="country", range_y=[0,4000000000])

fig = px.bar(newdf, x="Month", y="Elevation", color="Month",
             animation_frame="Year", range_y=[Mini-50, Maxi+50], width=1500, height=1000, title="Elevation of water at Lake Mead (in feet above sea level) at end of month by year")

fig.write_html("animationLakeMead.html")

fig.show()
```

```python
from cmath import nan
import pandas as pd
import numpy as np
import plotly.express as px
from datetime import datetime
from plotly import graph_objs as go

df = pd.read_excel("LakeMead_MonthlyElevationData.xlsx", index_col=0)

MonthList = ['JAN', 'FEB',	'MAR',	'APR',	'MAY',
             'JUN',	'JUL',	'AUG',	'SEP',	'OCT',	'NOV',	'DEC']
YearStart   = 1940
YearEnd     = 2021


newdf = pd.DataFrame(columns=['Date', 'Elevation'])
for year in range(YearStart, YearEnd+1):
    for mth in MonthList:
        Ele = df.at[year, mth]
        datestr = str(year) + '-' + mth + '-01'
        dt = datetime.strptime(datestr, "%Y-%b-%d")
        new_row = pd.Series({'Date': dt, 'Elevation': Ele})
        newdf = pd.concat([newdf, new_row.to_frame().T], ignore_index=True)

# print(newdf)

# fig = px.line(newdf, x='Date', y='Elevation')
# fig.show()




# Create figure to plotly it
fig = go.Figure()

fig.add_trace(
    go.Scatter(x=list(newdf.Date), y=list(newdf.Elevation)))

# Add range slider
fig.update_layout(    
    title="LakeMead ElevationData by year & Monthly",
    title_x=0.5,
    autosize=True,
    #width=1000,
    #height=600,
    xaxis=dict(
        # rangeselector=dict(
        #     buttons=list([
        #         dict(step="all")
        #     ])
        # ),

        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

# Interactive
#fig.write_html("applestock.html")

fig.show()

```

```python
# import plotly.express as px
from cmath import nan
import pandas as pd
import numpy as np
import plotly.graph_objects as go

df = pd.read_excel("LakeMead_LowHighElevationData.xlsx")

years = df.loc[:, 'Year'].tolist()
low_elv = df.loc[:, 'Low Elev'].tolist()
hi_elv = df.loc[:, 'High Elev'].tolist()


# low_date = df.loc[:, "Date"].tolist()
# print(low_date)

# the 1st "Date" for Low
low_date = df.iloc[:, 1].tolist()
# print(low_date)

# the 2nd "Date" for High
hi_date = df.iloc[:, 4].tolist()
# print(hi_date)


# the 1st "Time" for Low
low_time = df.iloc[:, 2].tolist()
# print(low_time)

# the 2nd "Time" for High
hi_time = df.iloc[:, 5].tolist()
# print(hi_time)


low_date_updated = []
index = 0

for dt in low_date:
    tm = low_time[index]
    hrs = tm // 100
    if hrs == 24:
        hrs = 0
    mins = tm % 100
    low_date_updated.append(dt.replace(
        year=years[index], hour=hrs, minute=mins))
    index += 1

# print(low_date_updated)

hi_date_updated = []
index = 0
for dt in hi_date:
    tm = hi_time[index]
    hrs = tm // 100
    if hrs == 24:
        hrs = 0
    mins = tm % 100
    hi_date_updated.append(dt.replace(
        year=years[index], hour=hrs, minute=mins))
    index += 1

# print(hi_date_updated)


# https://plotly.com/python/hover-text-and-formatting/#customizing-hover-text-with-a-hovertemplate
fig = go.Figure()
fig.add_trace(go.Scatter(x=low_date_updated, y=hi_elv, name='High'))
fig.add_trace(go.Scatter(x=hi_date_updated,  y=low_elv, name='Low'))
# fig = px.line(df, x="Year", y="Low Elev", title="xxxxx") #hover_data=["continent", "pop"]
# fig = px.line(df, x="Year", y="High Elev", title="xxxxx") #hover_data=["continent", "pop"]


# fig = px.line(df, x="year", y="lifeExp", color="country", title="layout.hovermode='x'", hover_data=["continent", "pop"])
fig.update_traces(mode="markers+lines")
fig.update_traces(hovertemplate='Date: %{x} <br>Elevation: %{y}')
# # fig.update_layout(hovermode="x")
fig.update_xaxes(showspikes=True)
fig.update_yaxes(showspikes=True)


fig.update_layout(title='The highest and lowest water elevation at Lake Mead (in feet above sea level) by year.',
                  xaxis_title='Year',
                  yaxis_title='Elevation')

fig.write_html("LowHighElevationData.html")

fig.show()
```
